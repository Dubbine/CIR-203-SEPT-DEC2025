```python
# 1. Create product dictionary
inventory = {
    "Laptop": 15,
    "Phone": 8,
    "Tablet": 12,
    "Headphones": 20,
    "Camera": 5
}

# 2. Add new product and update existing one
inventory["Smartwatch"] = 10
inventory["Phone"] = 14

# 3. Function to return products with stock < 10
def low_stock(inv):
    return {k: v for k, v in inv.items() if v < 10}

print("Low stock products:", low_stock(inventory))

# 4. Delete discontinued product
del inventory["Camera"]
print("Updated inventory:", inventory)

# 5. Use .items() to print all products and quantities
for product, qty in inventory.items():
    print(f"{product}: {qty}")
```
