```python
# 1. Create a list of 10 routes
routes = ["Nairobi-Kisumu", "Mombasa-Nakuru", "Nakuru-Eldoret", "Nairobi-Thika", 
          "Naivasha-Nakuru", "Nyeri-Meru", "Nairobi-Mombasa", "Nakuru-Kitale", 
          "Nanyuki-Nakuru", "Narok-Bomet"]

# 2. Append new route and remove discontinued one
routes.append("Kisii-Kericho")
routes.remove("Nakuru-Eldoret")

# 3. Sort alphabetically then reverse
routes.sort()
routes.reverse()

# 4. Count how many routes start with 'N'
count_N = len([r for r in routes if r.startswith("N")])
print("Routes starting with 'N':", count_N)

# 5. List comprehension: routes longer than 10 chars
long_routes = [r for r in routes if len(r) > 10]
print("Routes longer than 10 characters:", long_routes)
```
