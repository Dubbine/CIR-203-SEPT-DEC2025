```python
# 1. Create a patient tuple
patient = ("John Doe", 45, "120/80", 72)

# 2. Access age and heart rate
print("Age:", patient[1])
print("Heart Rate:", patient[3])

# 3. Why tuples are suitable
print("Tuples are immutable, meaning data like patient vitals cannot be accidentally changed, "
      "ensuring record integrity.")

# 4. Convert tuple to list, update, and convert back
patient_list = list(patient)
patient_list[3] = 75  # Update heart rate
patient = tuple(patient_list)
print("Updated patient record:", patient)

# 5. Tuple of 5 patients and extract all names
patients = (
    ("John Doe", 45, "120/80", 72),
    ("Jane Smith", 38, "110/70", 80),
    ("Paul Kim", 50, "130/85", 78),
    ("Mary Lee", 29, "115/75", 70),
    ("Lucy Wanjiru", 40, "125/80", 74)
)

names = [p[0] for p in patients]
print("All patient names:", names)
```
