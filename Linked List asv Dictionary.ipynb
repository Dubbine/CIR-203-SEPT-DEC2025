```python
# Linked list implemented as a dictionary of students

students = {}   # stores all student nodes


# Function to add a student node
def add_student(key, name, adm_no, grade1, grade2, grade3, next_key=None):
    students[key] = {
        "name": name,
        "adm_no": adm_no,
        "grades": {
            "CIR205": grade1,
            "CIT117": grade2,
            "MAS101": grade3
        },
        "next": next_key
    }


# Function to print the linked list
def print_students(start_key):
    current = start_key
    while current is not None:
        node = students[current]
        print(f"Student Key: {current}")
        print(f"  Name: {node['name']}")
        print(f"  Admission No: {node['adm_no']}")
        print(f"  Grades: {node['grades']}")
        print()
        current = node["next"]


# Create linked list nodes
add_student("S1", "DUBBINE SAM", "CIM/00117/024", "A", "B+", "A-", "S2")
add_student("S2", "MERVEILE DUFANDA", "CIS/0008/024", "B", "A", "A-", "S3")
add_student("S3", "CLAIVERT MOLIX", "CIT/0089/024", "A-", "B", "B+", None)

# Print the linked list
print_students("S1")
```
